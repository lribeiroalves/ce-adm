```python
def fibo(n:int) -> list:
    fibo = [1, 1]
    if n < 1:
        return [0]
    elif n < 2:
        return [1]
    elif n < 3:
        return fibo
    
    for _ in range(n-2):
        f = fibo[-1] + fibo[-2]
        fibo.append(f)
    
    return fibo
```
